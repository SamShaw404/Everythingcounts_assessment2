**Dataset Bias and limitations**

While this dataset provides valuable insights into musicians' experiences with burnout, several limitations and potential biases should be acknowledged:

- **Sample size** - The dataset has only 48 responses, which is relatively small. This reduces the generalisability of the findings and may not capture the full diversity of experiences among professional musicians.

- **Self-reported data** - The survey relied on self-reported data, which introduces the potential for inaccuracies as participants may have been influenced by Memory Bias, Social disirability bias and misinterpretaion of questions.

- **Cultural and Geographic Bias** - The dataset predominatly represents musicians based in the UK, which limits the ability to generalise findings from other regions, as cultural and systemic differences in the music industry may significantly influence experiences of burnout.

- **Demographic Representation** - Certain groups, such as higher-income musicians or those working outside traditional ensemble structures, are underrepresented. The imbalance may skew finding and underplay the experiences of less visible demographics within the profession.

# First Hypothesis - Financial Challenges

**Research Question**: Does the annual income signigiciantly predict burnout frquency among musicians?

**- Null Hypothesis (H₀)**: Annual income does not significantly predict burnout frequency.

**- Alternative Hypothesis (H₁)**: Annual income significantly predicts burnout frequency.

**Statistical test - Kruskal-Wallis Test**
This test was selcted because it is a non-parametric statistical test that compares the medians of two independent samples. 
I could not use a one way ANOVA as that would require the assumption of normality or equal variances, making it unsuitable for this small sized dataset.



```python
# Box plot visualising how burnout frequency varies acorss income levels.
plt.figure(figsize=(10, 6))
sns.boxplot(x='Income', y='Burnout_Freq', data=mb, showfliers=False, palette='coolwarm')
plt.title("Burnout Frequency Across Income Levels")
plt.xlabel("Income Levels (£)")
plt.ylabel("Burnout Frequency")
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import kruskal

# Group data by income
income_groups = mb.groupby('Income')['Burnout_Freq'].apply(list)

# Perform Kruskal-Wallis test
stat, p_value = kruskal(*income_groups)
print(f"Kruskal-Wallis Statistic: {stat}")
print(f"P-value: {p_value}")

# Interpret results
if p_value < 0.05:
    print("Reject null hypothesis: Significant differences in burnout frequency among income groups.")
else:
    print("Fail to reject null hypothesis: No significant differences in burnout frequency among income groups.")
```

**Hypothesis 1 Result**

The Kruskal-Wallis test produced a statistic (H) of 9.68 with a p-value of 0.469, indicating no statistically significant differences in burnout frequencies across income levels, despite the visible differences in medians. Therefore, we fail to reject the null hypothesis, suggesting that income is not a significant predictor of burnout frequency in this sample of musicians.


**Hypothesis 1 Discussion**

The Kruskal-Wallis test suggest that annual income does not significantly predict burnout frequency among musicians. Despite financial challenges being a frequetly cited stressor in the music profession, the data did not show significaant variation in burnout frequency across income categories (Längle, 2003; Pines and Aronson, 1988).

The data had uneven group sizes, where both the higher income categories and the burnout frequency had very few observations, reducing the statistical power to detect meaningful differences, and therefore masking potential trends in the data.

**Future research directions** Future studies should aim for larger, more balanced datasets with greater representation in higher income, with additional variables like workload variability, access to mental health resources, and coping strategies should be included to provide a more comprehensive understanding of burnout.



# Hypothesis 2 - Perfectionism

**Research Question**: Is there a significant difference in burnout frequency between perfectionists and non-perfectionists?

**Null Hypothesis (H₀)**: Burnout frequency does not differ significantly between perfectionists and non-perfectionists.

**Alternative Hypothesis (H₁)**: Burnout frequency differs significantly between perfectionists and non-perfectionists.

**Statistical test - T-tes**
The T test compares the means of two independant groups, as the Perfectionis group is a binary variable, and burnout_freq is treated as a continuous dependent variable, under the assumption of normality.

```python
# The barplot compares average burnout frequences for perfectionists and non-perfectionists.
# Set seaborn style
sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.barplot(x='Perfectionist', y='Burnout_Freq', data=mb, ci=None, palette='muted')
plt.title("Burnout Frequency by Perfectionism", fontsize=16)
plt.xlabel("Perfectionist (1 = Yes, 0 = No)", fontsize=12)
plt.ylabel("Burnout Frequency", fontsize=12)
plt.xticks([0, 1], ["No", "Yes"])  
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import ttest_ind

# T-test for perfectionists vs non-perfectionists
Perfectionists = mb[mb['Perfectionist'] == 1]['Burnout_Freq']
Non_perfectionists = mb[mb['Perfectionist'] == 0]['Burnout_Freq']
t_stat, p_val = ttest_ind(Perfectionists, Non_perfectionists)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

**Hypothesis 2 Result**

The P value (0.375) which is greater than 0.05, indicating that there is no statistically significant difference in burnout frequency between perfectionists and non-perfectionists. This, we fail to reject the null hypothesis.


**Hypothesis 2 discussion**

The results suggest that perfectionism does not sognificantly affect burnout frequwncy among musicians in this sample. However, these results contrast with the broader psychological literature, which identifies perfrectionism as a key driver of stress and burnout. Stoeber and Otto (2006) explore perfectionistic strivings which are associated with motivation and self improvement, where Stoeber contunes his research (Stoeber and Eismann, 2007) and explores how this often leads to self-doubt, emotional exhaustion and then burnout.

The lack of significance in this dataset may refect the diverse impacts of perfectionism, where some individuals may mitigate its effects through structed practices and resilience (Längle, 2004; Hamachek, 1978). Cultural and systemic influence also play a role, as Curran and Hill (2016) observed increasing perfectionism in younger generations, heighteing day to day stress levels. Powell (2021) explores how competitive norms in education perpetiate perfectionistic concerns and exacerbate burnout in marginalised groups.

**Future research directions**
Future studies should expand the sample size and diversity to capture a wider range of experiences and better understand the naunced relationship between perfectionism and burnout. In a longitudional study, examining how perfectionism influences burnout over time, patticularly at different career stages.


# Hypothesis 3 - Workload Dynamics

**Research Question**: Does the frequency of working with new ensembles correlate with burnout frequency?

**Null Hypothesis (H₀)**: New ensemble frequency does not correlate with burnout frequency.

**Alternative Hypothesis (H₁)**: New ensemble frequency correlates with burnout frequency.

**Statistical test - Spearman correlation** As both brunout_freq and new_ensemble_frew are do not follow a normal distribution,. The test evaluates whether a monotonic relationship exists between the two variables. This is important when the relationship may increase or decrease consistently but not at a constant rate.


```python
# Scatterplot to show the relationship between new ensemble frequency and burnout frequency
sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.scatterplot(x='New_ensemble_freq', y='Burnout_Freq', data=mb, color='blue', s=50)
sns.regplot(x='New_ensemble_freq', y='Burnout_Freq', data=mb, scatter=False, color='red', ci=None)
plt.title("Burnout Frequency vs New Ensemble Frequency", fontsize=16)
plt.xlabel("New Ensemble Frequency")
plt.ylabel("Burnout Frequency")
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import spearmanr

# Spearman correlation
corr, p_value = spearmanr(mb['New_ensemble_freq'], mb['Burnout_Freq'])
print(f"Spearman correlation: {corr}, P-value: {p_value}")
```

**Hypothesis 3 result**

The spearman correlation coefficient suggests a weak positive relationship between the frequency of working with new ensembkes and burnout frequencey. However the p value (0.165) is greater than 0.05, indicating that the observed correlation is not statistically significant. Therefore we fail to reject the null hypothesis, suggestin no significant correlation between new ensemble frequency and burnout frequency in this sample.

**Hypothesis 3 discussion**

The results suggest that the frequency of working with new ensembles does not significiantly correlate with burnout freuqncy among musicians in this sample. Although the weak positive correlation (p=0.204) implies that the higher ensemble turnover might slightly increase burnout, the lack of statistical significiance indiciates that this relationship is not robust.

This contasts with existing literature hightling workload variability as a stressor.  Längle (2003) and Portí et al. (2020) note that frequent professional changes, such as adapting to new ensembles, can disrupt a musician's sense of control (inner-consent) and lead to increased stress. 

The dataset may lack sufficient representation of musicians experiencing high ensemble turnover, limiting the power to detect a significant relationship.

**Future research directions** Future studies should expand the sample size and diversity to improve the reliability of results and caputure a wider range of ensemble turnover experience.



# Hypothesis 4 - Educational Background

**Research Question**: Do musicians who attended conservatoires experience burnout more frequently than those who did not?

**Null Hypothesis (H₀)**: Conservatoire attendance does not significantly affect burnout frequency.

**Alternative Hypothesis (H₁)**: Conservatoire attendance significantly affects burnout frequency.

**Statistical test - Mann-Whitney U Test** This is a non-parametic test, to compare two independent groups, evaluating whether the distributionof burnout freuqnecy differs significantly between conservatoire and non-conservatoire groups without assuming normality.



```python
# Boxplot 
# Set seaborn style
sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.boxplot(x='Conservatoire', y='Burnout_Freq', data=mb, showfliers=False, palette='pastel')
plt.title("Burnout Frequency by Conservatoire Attendance", fontsize=16)
plt.xlabel("Conservatoire Attendance (1 = Yes, 0 = No)")
plt.ylabel("Burnout Frequency")
plt.xticks([0, 1], ["No", "Yes"])  
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import mannwhitneyu

# Mann-Whitney U test
group1 = mb[mb['Conservatoire'] == 1]['Burnout_Freq']
group2 = mb[mb['Conservatoire'] == 0]['Burnout_Freq']
u_stat, p_val = mannwhitneyu(group1, group2)
print(f"U-statistic: {u_stat}, P-value: {p_val}")
```

**Hypothesis 4 result**

The P value is 0.304 is greater than 0.05, indicating no staistically significant difference in burnout frequency between musicians who attended conservatoire and those who did not. Therefore, we fail to reject the null hypothesis.

**Hypothesis 4 discussion**
The reaults suggest that conservatoire attendance does not significantly include burnout freuqnecy among musicians in this sample. The findings contrast with prior research emphasising the high-pressure encironoment of conservatoires. Presland(2005) describes conservatoire as intense and demanding settings that often amplify stress and anzity in students. Similarly, Zabuska (2017) highlights the importance of supportive teacher-student relationships, noting that their absence in such high-stress environments can exacerbate burnout.

Musicians who attend conservatoire may develop resilience and coping strateges that offset the stress associated with such environments, as well as the normalised highs tress levels as part of the conservatoire culutrue, making them less likely to report burnout (Powell, 2021).


**Future research directions**
Furture research should collect data from a larger and more diverse population of musicians to bettwe capture the effects of conservatoire training on burnout, and explore how attenance interacts with other variables.
